```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import sklearn
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from urllib.request import urlretrieve
```

BMI (body mass index)


```python
fig = px.histogram(data,
                       x = 'bmi',
                       marginal ='box',
                       color_discrete_sequence = ['red'],
                       title = 'BMI Distribution')
fig.update_layout(bargap=0.1)
fig.show()
```

Charges and sex compared


```python
fig = px.histogram(data,
                   x = 'charges',
                   marginal = 'box',
                   color = 'sex',
                   color_discrete_sequence = ['red', 'blue'],

                   title = 'Charges Distribution by Sex'
                   )
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.box(data,
                   x = 'charges',
                   color = 'sex',
                   color_discrete_sequence = ['red', 'blue'],

                   title = 'Charges Distribution by Sex'
                   )
fig.show()
```

```python
fig = px.histogram(data,
                   x='charges',
                   marginal = 'box',
                   color = 'smoker',
                   color_discrete_sequence=['green',"gray"],
                   title = 'Anuual medical charges')
fig.update_layout(bargap=.2)
fig.show()
```

```python
data.region.unique()
```

```python
fig = px.histogram(data,
                   x='charges',
                   marginal = 'box',
                   color = 'region',
                   color_discrete_sequence=['green',"gray",'red','blue'],
                   title = 'Anuual medical charges')
fig.update_layout(bargap=.2)
fig.show()
```

```python
fig = px.scatter(data,
                 x='age',
                 y = 'charges',
                 color = 'smoker',
                 opacity = 0.8,
                 hover_data=['sex'],
                 title = 'Age vs charge')
fig.update_traces(marker_size=5)
fig.show()
```

We can observe 3 trenlines forming on the graph, all 3 with positive correlations, one for people who dont smoke, and 2 for people who smoke, we can assume that 2 lines are formed, one for smokers with low BMI and one with high BMI

**BMI and charges**

```python
fig = px.scatter(data,
                 x='bmi',
                 y = 'charges',
                 color = 'smoker',
                 opacity = 0.8,
                 hover_data=['sex'],
                 title = 'BMI vs charge')
fig.update_traces(marker_size=5)
fig.show()
```

We can observe in smokers, when BMI>30 charges jumps from minimun 18k to minimum 32k
